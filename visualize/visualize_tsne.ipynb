```python
import numpy as np
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
palette_aaai_2024 = ["#5386e4","#5c946e","#694873","#ff2c55","#ffca3a"]
plt.rc('font',family='Times New Roman')
```

```python
# originally saved in output/subfolder/tsne/****tsne.csv
# format: x, y, type
df_normal = pd.read_csv(os.path.join(data_dir, f'ndcg@10(all)_tsne_1v2_new_normal.csv'))
df_inter_v3 = pd.read_csv(os.path.join(data_dir, f'ndcg@10(all)_tsne_1v2_new_inter_v3.csv'))
df_intra = pd.read_csv(os.path.join(data_dir, f'ndcg@10(all)_tsne_1v2_new_intra.csv'))
df_fair_v3 = pd.read_csv(os.path.join(data_dir, f'ndcg@10(all)_tsne_1v2_new_fair_v3.csv'))
```

```python
fig, ax =plt.subplots(2,2,constrained_layout=True, figsize=(16, 8),sharex=True,sharey=False)

plt.subplot(2, 2, 1)
axesSub = sns.scatterplot(data=df_normal, x='x', y='y', hue='type', 
                          palette=palette_aaai_2024)
handles, labels = ax[0][0].get_legend_handles_labels()
ax[0][0].legend(handles=handles , labels=labels)
ax[0][0].set_title('Normal', fontdict={'family' : 'Times New Roman', 'size'   : 32, 'weight': 'bold'})
ax[0][0].set_xticks([])
ax[0][0].set_yticks([])
ax[0][0].set_ylabel("")

plt.subplot(2, 2, 2)
axesSub = sns.scatterplot(data=df_inter_v3, x='x', y='y', hue='type',
                          palette=palette_aaai_2024)
handles, labels = ax[0][1].get_legend_handles_labels()
ax[0][1].legend(handles=handles , labels=labels)
ax[0][1].set_title('Inter', fontdict={'family' : 'Times New Roman', 'size'   : 32, 'weight': 'bold'})
ax[0][1].set_xticks([])
ax[0][1].set_yticks([])
ax[0][1].set_ylabel("")

plt.subplot(2, 2, 3)
axesSub = sns.scatterplot(data=df_intra, x='x', y='y', hue='type',
                          palette=palette_aaai_2024)
handles, labels = ax[1][0].get_legend_handles_labels()
ax[1][0].legend(handles=handles , labels=labels)
ax[1][0].set_title('Intra', fontdict={'family' : 'Times New Roman', 'size'   : 32, 'weight': 'bold'})
ax[1][0].set_xticks([])
ax[1][0].set_yticks([])
ax[1][0].set_ylabel("")

plt.subplot(2, 2, 4)
axesSub = sns.scatterplot(data=df_fair_v3, x='x', y='y', hue='type',
                          palette=palette_aaai_2024)
handles, labels = ax[1][1].get_legend_handles_labels()
ax[1][1].legend(handles=handles , labels=labels)
ax[1][1].set_title('II-GOOT', fontdict={'family' : 'Times New Roman', 'size'   : 32, 'weight': 'bold'})
ax[1][1].set_xticks([])
ax[1][1].set_yticks([])
ax[1][1].set_ylabel("")

legend_font = {
    'family': 'Times New Roman',  # 
    'size': 24,  # 
}

lines, labels = ax[0][0].get_legend_handles_labels()
fig.legend(lines, labels, loc = 'lower center', ncol=2, prop=legend_font,bbox_to_anchor=(0.5, -0.1)) # bbox_to_anchor=(0.5, 0.92),
ax[0][0].get_legend().remove()
ax[0][1].get_legend().remove()
ax[1][0].get_legend().remove()
ax[1][1].get_legend().remove()

# sns.move_legend(ax[0][0], "lower left", prop=legend_font)
# sns.move_legend(ax[0][1], "lower left", prop=legend_font)
# sns.move_legend(ax[1][0], "lower left", prop=legend_font)
# sns.move_legend(ax[1][1], "lower left", prop=legend_font)


plt.tight_layout()
# plt.savefig(os.path.join(data_dir, f"all_tsne_1v2.png"))
```
