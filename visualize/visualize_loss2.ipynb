```python
import pandas as pd
import numpy as np
import seaborn as sns
# import palettable
import os 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Times New Roman')
```

```python
import pickle
dir_name = "/home/lm/recommendation/OT_fairness2/myCode_ot_final/visualize/loss"

# originally saved in output/subfolder/loss/loss.pkl
# format: dict {epoch: (overall_loss, original_loss, ot_loss_inter, ot_loss_intra)}
loss_beauty_fair = pickle.load(open(os.path.join(dir_name, "loss_beauty_fair_e200.pkl"), 'rb'))
loss_grocery_normal = pickle.load(open(os.path.join(dir_name, "loss_grocery_normal_e200.pkl"), 'rb'))
loss_grocery_fair = pickle.load(open(os.path.join(dir_name, "loss_grocery_fair_e200.pkl"), 'rb'))
loss_health_normal = pickle.load(open(os.path.join(dir_name, "loss_health_normal_e200.pkl"), 'rb'))
loss_health_fair = pickle.load(open(os.path.join(dir_name, "loss_health_fair_e200.pkl"), 'rb'))
loss_beauty_normal = pickle.load(open(os.path.join(dir_name, "loss_beauty_normal_e200.pkl"), 'rb'))
```

```python
loss_beauty_fair_inter = [loss_beauty_fair[epoch][2] for epoch in range(200)]
loss_beauty_normal_inter = [loss_beauty_normal[epoch][2] for epoch in range(200)]
loss_grocery_fair_inter = [loss_grocery_fair[epoch][2] for epoch in range(200)]
loss_grocery_normal_inter = [loss_grocery_normal[epoch][2] for epoch in range(200)]
loss_health_fair_inter = [loss_health_fair[epoch][2] for epoch in range(200)]
loss_health_normal_inter = [loss_health_normal[epoch][2] for epoch in range(200)]
```

```python
df_beauty = pd.DataFrame({
    'original': loss_beauty_normal_inter,
    "II-GOOT": loss_beauty_fair_inter,
})
df_grocery = pd.DataFrame({ 
    'original': loss_grocery_normal_inter,
    "II-GOOT": loss_grocery_fair_inter,
})
df_health = pd.DataFrame({
    'original': loss_health_normal_inter,
    "II-GOOT": loss_health_fair_inter,
})
```

```python

palette_aaai_2024 = ["coral", "cornflowerblue"]
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 16/3), dpi=400)
sns.set_style("ticks")

legend_font = {
    'family': 'calibri', 
    'size': 19,  
}


####################
### Beauty
####################
sns.lineplot( data=df_beauty, 
             ax=axes[0],
             dashes=False,
             linewidth=2,
             palette=palette_aaai_2024)

axes[0].set_title("Beauty", fontdict={'family' : 'calibri', 'size'   : 32, 'weight': 'bold'})
axes[0].set_ylabel('$L_{inter}$', fontdict={'family' : 'calibri', 'size'   : 32, 'weight': 'bold'})
axes[0].set_xlabel('epoch', fontdict={'family' : 'calibri', 'size'   : 32, 'weight': 'bold'})

####################
### grocery
####################
sns.lineplot( data=df_grocery, 
             ax=axes[1],
             dashes=False,
             linewidth=2,
             palette=palette_aaai_2024)

axes[1].set_title("Grocery", fontdict={'family' : 'calibri', 'size'   : 32, 'weight': 'bold'})
axes[1].set_ylabel('$L_{inter}$', fontdict={'family' : 'calibri', 'size'   : 32, 'weight': 'bold'})
axes[1].set_xlabel('epoch', fontdict={'family' : 'calibri', 'size'   : 32, 'weight': 'bold'})


####################
### Health
####################
sns.lineplot( data=df_health, 
             ax=axes[2],
             dashes=False,
             linewidth=2,
             palette=palette_aaai_2024)

axes[2].set_title("Health", fontdict={'family' : 'calibri', 'size'   : 32, 'weight': 'bold'})
axes[2].set_ylabel('$L_{inter}$', fontdict={'family' : 'calibri', 'size'   : 32, 'weight': 'bold'})
axes[2].set_xlabel('epoch', fontdict={'family' : 'calibri', 'size'   : 32, 'weight': 'bold'})

sns.move_legend(axes[2], "upper left",  prop=legend_font)
sns.move_legend(axes[1], "upper left",  prop=legend_font)
sns.move_legend(axes[0], "upper left",  prop=legend_font)

# 调整刻度轴字体大小
axes[0].tick_params(axis='x',labelsize=16,) 
axes[0].tick_params(axis='y',labelsize=16,) 
axes[1].tick_params(axis='x',labelsize=16,) 
axes[1].tick_params(axis='y',labelsize=16,) 
axes[2].tick_params(axis='x',labelsize=16,) 
axes[2].tick_params(axis='y',labelsize=16,)


plt.tight_layout()

```
